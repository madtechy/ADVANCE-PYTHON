```python
from dataclasses import dataclass
```

```python
@dataclass
class Triangle:
    a: int | float
    b: int | float
    c: int | float

    def __post_init__(self):
        errors = []

        if not isinstance(self.a, (int, float)) or self.a <= 0:
            errors.append("Side a should be a positive integer or float")
        if not isinstance(self.b, (int, float)) or self.b <= 0:
            errors.append("Side b should be a positive integer or float")
        if not isinstance(self.c, (int, float)) or self.c <= 0:
            errors.append("Side c should be positive integer or float")
        if not self.is_triangle():
            errors.append("Sum of amy two sides should be always greater than third")

        if errors:
            raise ValueError("\n".join(errors))

    def is_triangle(self):
        if (
            self.a + self.b > self.c and
            self.b + self.c > self.a and
            self.c + self.a > self.b
        ):
            return True
        else:
            return False
        
    def perimeter(self):
        return self.a + self.b + self.c
        
    def area(self):
        s = self.perimeter() / 2
        a = (s * (s - self.a) * (s - self.b) * (s - self.c))**(1/2)
        return a
```

```python
class India:

    def capital(self):
        print("New Delhi is capital of India")

    def language(self):
        print("Hindi is widely spoken in India")    
```

```python
class USA:
    def capital(self):
        print("Washington D.C is capital of USA")

    def language(self):
        print("English is widely spoken in USA")    
```

```python
class France:
    def capital(self):
        print("Paris is capital of France")

    def language(self):
        print("French is widely spoken in France")
```

### Calculating Area and perimeter for different shapes

```python
@dataclass
class Rectangle:
    width: int|float
    height: int|float

    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def area(self):
        return self.width * self.height
```

### 

```python
from math import pi
```

```python
pi
```

```python
@dataclass
class Circle:
    radius: int|float

    def perimeter(self):
        return 2 * pi * self.radius
    
    def area(self):
        return pi * (self.radius ** 2)
```

```python
s1 = Triangle(12, 13, 14)
s2 = Rectangle(20, 10)
s3 = Circle(14)
s4 = Triangle(3, 4, 5)
s5 = Circle(21)
s6 = Rectangle(5, 3)
```

```python
shapes = [s1,s2, s3, s4, s5, s6]
```

```python
for i in shapes:
    print(i)
    print(type(i))
    print(f"Perimeter : {i.perimeter():.4f}")
    print(f"Area : {i.area():.4f}")
    print("\n=================================================\n")
```
